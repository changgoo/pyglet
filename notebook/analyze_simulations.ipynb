# Inspect simulation data using yt

```python
import yt
```

```python
# https://yt-project.org/doc/cookbook/complex_plots.html#multipanel-with-axes-labels
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import AxesGrid

fig = plt.figure(figsize=(15,10))

# See http://matplotlib.org/mpl_toolkits/axes_grid/api/axes_grid_api.html
# These choices of keyword arguments produce a four panel plot that includes
# four narrow colorbars, one for each plot.  Axes labels are only drawn on the
# bottom left hand plot to avoid repeating information and make the plot less
# cluttered.
grid = AxesGrid(
    fig,
    (0.075, 0.075, 0.85, 0.85),
    nrows_ncols=(2, 3),
    axes_pad=1.0,
    label_mode="1",
    share_all=True,
    cbar_location="right",
    cbar_mode="each",
    cbar_size="3%",
    cbar_pad="0%",
)

for i,d in enumerate(dirs):
    sim = LoadSim(d)
    ds = sim.load_athdf(num=520, output_id=2, load_method="yt")
    slc = yt.SlicePlot(ds,'z','density')
    slc.annotate_magnetic_field()
    plot = slc.plots['density']
    plot.figure = fig
    plot.axes = grid[i].axes
    plot.cax = grid.cbar_axes[i]
    slc.render()
```

# Creating Xray mocks

```python
sim = LoadSim(dirs[0])
ds = sim.load_athdf(num=520, output_id=2, load_method="yt")
```

```python
xray = Xray(
    ds,
    emin=0.1,
    emax=3.0,
    binscale="linear",
    nbins=100,
    savdir=os.fspath(sim.basedir),
    dist_kpc = 50,
)
```

```python
xray.set_regions()
```

```python
ax = "z"
```

```python
xray.set_filenames(axis=ax, photon=True, event=True, simput=False)
```

```python
xray.create_simput(axis=ax, overwrite=False)
```

```python
for inst in ["lem_2eV", "lem_0.9eV"]:
    for exp in [10, 100]:
        xray.instrument_simulator(exp=exp, inst=inst, overwrite=False)
```

```python
fig = xray.show(
        match=dict(dist="50kpc", axis="z", exptime="100ks", inst="lem_2eV"),
        vmax=None,
        vmin=None,
    )
```

# Creating Joint PDFs of LOS velocity and temperature

```python
xray.add_xray_fields()
```

```python
profile = xray.create_profile(xy="T-vz")
```

```python
dset = xray.convert_profile_to_dataset(profile)
```

```python
from matplotlib.colors import LogNorm
```
